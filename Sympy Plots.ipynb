```python
import numpy as np
from matplotlib import pyplot as plt
from math import pi

u=0     #x-position of the center
v=1.   #y-position of the center
a=2.     #radius on the x-axis
b=4.    #radius on the y-axis

t = np.linspace(0, 2*pi, 100)
plt.figure()
ax = plt.gca()
ax.plot( u+a*np.cos(t) , v+b*np.sin(t) )
ax.grid(color='lightgray',linestyle='--')
#ax.set_title(r'$\frac{(x+3)^2}{9}+\frac{(y-5)^2}{3}=1$' + '\n', fontsize=22)
plt.show()
```
